## Unnamed:0 should be dropped

```python
df.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
```

## Salary Parsing

```python
# Salary contains -1
df=df[df['Salary Estimate']!='-1']
```

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.split("(")[0])
```

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.replace("$","").replace("K",""))
```

```python
df['hourly']=df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)
```

```python
df['employer_provided']=df['Salary Estimate'].apply(lambda x:1 if 'employer provided' in x.lower() else 0)
```

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.lower().replace('per hour',''))
```

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.lower().replace('employer provided salary:',''))
```

```python
df['Salary Estimate']=df['Salary Estimate'].str.strip()
```

```python
df['min_salary']=df['Salary Estimate'].apply(lambda x: int(x.split('-')[0]))
```

```python
df['max_salary']=df['Salary Estimate'].apply(lambda x: int(x.split('-')[1]))
```

```python
df['avg_salary']=(df['min_salary']+df['max_salary'])/2
```

```python
# Hourly wage to annual
df['min_salary']=df.apply(lambda x: x.min_salary*2 if x.hourly==1 else x.min_salary,axis=1)
df['max_salary']=df.apply(lambda x: x.max_salary*2 if x.hourly==1 else x.max_salary,axis=1)
```

```python
df[df.hourly==1][['Salary Estimate','hourly','min_salary','max_salary']]
```

## Company name text only

```python
df['Company Name']=df['Company Name'].apply(lambda x: x.split("\n")[0])
```

## State Feilds

```python
df['job_state']=df['Location'].apply(lambda x: x.split(",")[1])
```

```python
df['same_state']=df.apply(lambda x:1 if x.Location==x.Headquarters else 0,axis=1)
```

```python
df['job_state'] = df['job_state'].apply(lambda x: x.strip().lower() if x.strip().lower() != 'los angeles' else 'ca')
```

## Age of Company

```python
df['age']=df['Founded'].apply(lambda x: x if x<1 else 2024-x)
```

## Parse the Job Description

```python
df['python']=df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
df['sql']=df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
```

```python
df['excel']=df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df['tableau']=df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
```

```python
df["desc_length"]=df['Job Description'].apply(lambda x: len(x))
```

## Cleaning Job Title Column

```python
df['Job Title'].value_counts()
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
```

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'sr.' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'junior'
    else:
        return 'na'
```

```python
df['job_simplified']=df['Job Title'].apply(title_simplifier)
```

```python
df.job_simplified.value_counts()
```

```python
df['seniority']=df['Job Title'].apply(seniority)
```

```python
df.seniority.value_counts()
```

## Competitor Count

```python
df['num_comp']=df['Competitors'].apply(lambda x: len(x.split(",")) if x!='-1' else 0)
```
